```python
import pandas as pd
df = pd.read_csv('data/SPA_DataCatalogue_250822_NDSAP-aligned_updated.csv', dtype={'date__earliest': 'Int64', 'date__latest': 'Int64'})
df.head()
```

```python
# import dataset kpi linkage from xlsx
df_kpi = pd.read_excel('data/SPA_DataCatalogue_250611_KPIs.xlsx', sheet_name='Catalogue', engine='openpyxl')
df_kpi.rename(columns=
  {
  'ki_supplyofpublictransportvehicles':'A01', 
  'ki_publictransportaccessibility':'A02',
  'ki_walkability':'A03',
  'ki_coveragebycyclingfacilities':'A04', 
  'ki_coverageofthedrainagesystem':'A05',
  'ki_traineddrivers':'A06',
  'ki_transportrelatedinvestment':'A07', 
  'ki_localtrafficflow':'A08', 
  'ki_organisedroaduse':'A09', 
  'ki_parkingmanagement':'A10', 
  'ki_reductionofprivatevehicles':'B01',
  'ki_motorvehicletrafficreduction':'B02',
  'ki_declineinaccidents':'B03', 
  'ki_walkingandcycling':'B04',
  'ki_publictransportridership':'B05',
  'ki_lastmileuseofintermediatepublictransport':'B06', 
  } , inplace=True)

kpi_columns = ['A01', 'A02', 'A03', 'A04', 'A05', 'A06', 'A07', 'A08', 'A09', 'A10', 'B01', 'B02', 'B03', 'B04', 'B05', 'B06']
df_kpi['linked_to_kpi'] = df_kpi.apply(lambda row: ','.join([kpi for kpi in kpi_columns if row[kpi] == 1]), axis=1)
df_kpi.head()
```

```python
# merge both datasets by id
df = pd.merge(df, df_kpi[['id','linked_to_kpi']], on='id', how='left')
df.head(20)
```

```python
# Change case for certain columns
df['group'] = df['group'].str.title()
df['sector'] = df['sector'].str.title()
df['frequency'] = df['frequency'].str.capitalize()
df['granularity__spatial'] = df['granularity__spatial'].str.capitalize()
df['granularity__temporal'] = df['granularity__temporal'].str.capitalize()
df['access_type'] = df['access_type'].str.capitalize()
df['format'] = df['format'].str.capitalize()

# Fix 'Cd block'=> 'CD Block'
df['granularity__spatial'] = df['granularity__spatial'].str.replace('Cd block / town', 'CD Block / Town', regex=False)

# Fix Format 
df['format'] = df['format'].str.replace('Csv', 'CSV', regex=False)
df['format'] = df['format'].str.replace('Xls/xlsx', 'XLS/XLSX', regex=False)
df['format'] = df['format'].str.replace('Json', 'JSON', regex=False)
df['format'] = df['format'].str.replace('Zip', 'ZIP', regex=False)
df['format'] = df['format'].str.replace('Geotiff image', 'GeoTIFF image', regex=False)
df['format'] = df['format'].str.replace('Shapefile', 'ESRI Shapefile', regex=False)
df.head()
```

```python
# Clean up non-working URLs (to Null)
df['url'] = df['url'].replace('Jaipur City Mobility Plan', 'https://cuts-ccier.org/pdf/jaipur-city-mobility-plan.pdf')
df['url'] = df['url'].replace('Document not yet published', None)
df['access_method'] = df['access_method'].replace('Not currently available', None)
df['frequency'] = df['frequency'].replace('Unknown', None)
```

```python
# add url_download column for one row
df.loc[df.id == 290, 'url_download'] = 'https://orca.casa.ucl.ac.uk/~jens/sparc-dash/download/census2011.zip'
df[df.id == 290]
```

```python
# Add column lastupdated_group into last 2 years, 5 years, 10 years, etc.
import datetime
current_year = datetime.datetime.now().year
def lastupdated_group(year, yearlist=[2,5,10,15,20]):
    if year == 'N/A':
        return 'N/A'
    counter = 0
    for i in yearlist:
        if current_year - year <= i:
            return f'Last {i} yrs'
        counter += 1
    return f'{yearlist[counter-1]+1} yrs or older'
  
df['lastupdate'] = df['date__latest'].apply(lastupdated_group)                           
df.head()
```

```python
# Put keywords column to the back
keywords = df.pop('keywords')
df['keywords'] = keywords
df.head()
```

```python
df.to_json('../../public/data/datacatalogue.json', orient='records', indent=4)
```
