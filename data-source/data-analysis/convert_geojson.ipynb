```python
import geopandas as gpd
gdf = gpd.read_file("Jaipur_District_Block/Jaipur_District_Block.shp", driver="ESRI Shapefile", crs="EPSG:4326")
gdf.columns
```

```python
gdf = gdf[['ID_0', 'ISO', 'NAME_0', 'ID_1', 'NAME_1', 'ID_2', 'NAME_2', 'ID_3',
       'NAME_3', 
       'geometry']]
gdf.head()
```

```python
# Add color column, Generate random hex color values
import numpy as np
np.random.seed(42)

gdf['color'] = ['#' + ''.join(np.random.choice(list('0123456789ABCDEF'), size=6)) for _ in range(len(gdf))] 
gdf.head()
```

```python
gdf.to_file("../../public/geojson/placeholder.geojson", driver="GeoJSON")
```
